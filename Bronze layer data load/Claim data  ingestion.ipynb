```python
from pyspark.sql.functions import lit
claim_schema = "claim_id integer, policy_id integer, date_of_claim timestamp, claim_amount double, claim_status string, LastUpdatedTimeStamp timestamp"
df = spark.read.parquet("/mnt/landing/ClaimData", schema = claim_schema)
df_with_flag = df.withColumn("merge_flag", lit(False))
df_with_flag.write.option("path", "/mnt/bronze/Claim").mode("append").saveAsTable("bronze.Claim")
```

```python
from datetime import datetime
def getFilepathwithDate(filePath):
    current_time = datetime.now().strftime("%Y-%m-%d")
    return f"{filePath}/{current_time}"
```
