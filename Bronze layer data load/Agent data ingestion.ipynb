```python
from pyspark.sql.functions import lit

schema = "agent_id integer, agent_name string, agent_email string branch_id integer, create_timestamp timestamp"
df = spark.read.parquet('/mnt/landing/AgentData', schema=schema)
df_with_flag = df.withColumn("merge_flag", lit(False))  #to check if the file is processed or not
df_with_flag.write.option("path", "/mnt/bronze/Agent").mode("append").saveAsTable("bronze.Agent") #Save the agent data to bronze in delta format and also to bronze.Agent external table
```

```python
from datetime import datetime
def getFilepathwithDate(filePath):
    current_time = datetime.now().strftime("%Y-%m-%d")
    return f"{filePath}/{current_time}"
```
